### Build a Better Logistic Regression Model

```python
#make sure I have the right columns in data_clean dataframe
data_clean.head()
```

```python
#check data types
data_clean.dtypes
```

```python
#name feautures and target
columns = ["age", "job", "marital", "education", "default","housing", "loan", "contact", "month", "day_of_week",
          "campaign ", "pdays", "previous", "poutcome", "emp.var.rate","cons.price.idx", "cons.conf.idx", 
           "euribor3m", "nr.employed"]

target = ["y"]
```

```python
#split data into training and testing

# Create our features - drop target column
X = data_clean.drop(columns='y')

#clean features
X = pd.get_dummies(X)

# Create our target
y = data_clean.loc[:, target].copy()
X.head()
```

#### Check if there is class imbalance to determine if resampling techniques are necessary

```python
# Check the balance of our target values
y["y"].value_counts()

##We have imbalanced data.  Have significantly more observations in the class 'no'.  Therefore we apply different
#resampling techniques before training the model with logistic regression
```

```python
#split data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

#### Resampling technique #1: Resample the training data with the Random Oversampler

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=1)
X_resampled, y_resampled = ros.fit_resample(X_train, y_train)

Counter(y_resampled)
```

```python
# Train the Logistic Regression model using the resampled data
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resampled, y_resampled)
```

```python
# Calculate the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score
y_pred = model.predict(X_test)
print(balanced_accuracy_score(y_test, y_pred))
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, y_pred)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test, y_pred))
```

#### Resampling technique #2: Resample the training data with the SMOTE OVERSAMPLING

```python
X_resampled, y_resampled = SMOTE(random_state=1,sampling_strategy='auto').\
                            fit_resample(X_train, y_train)
Counter(y_resampled)
```

```python
# Train the Logistic Regression model using the resampled data
model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resampled, y_resampled)
```

```python
# Calculate the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score
y_pred = model.predict(X_test)
print(balanced_accuracy_score(y_test, y_pred))
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, y_pred)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test, y_pred))
```

#### Resampling technique #3: Resample the training data using the ClusterCentroids resampler

```python
cc = ClusterCentroids(random_state=1)
X_resampled, y_resampled = cc.fit_resample(X_train, y_train)
Counter(y_resampled)
```

```python
# Train the Logistic Regression model using the resampled data
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resampled, y_resampled)
```

```python
# Calculated the balanced accuracy score
y_pred = model.predict(X_test)
print(balanced_accuracy_score(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])
cm_df
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test, y_pred))
```

#### Resampling technique #4: Combination (Over and Under) Sampling with SMOTEENN

```python
smote_enn = SMOTEENN(random_state=0)
X_resampled, y_resampled = smote_enn.fit_resample(X, y)
Counter(y_resampled)
```

```python
model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resampled, y_resampled)
```

```python
# Calculated the balanced accuracy score
y_pred = model.predict(X_test)
print(balanced_accuracy_score(y_test, y_pred))
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, y_pred)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, y_pred))
```
